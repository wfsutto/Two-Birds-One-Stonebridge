# Exploratory Analysis of Ames Housing Dataset

#### Libraries

```python
import pandas as pd
```

## Load the Data

```python
ames = pd.read_csv('../data/AmesHousing.csv')
```

## Descriptive Statistics

```python
# Basic Info
ames.info()
```

```python
# Numeric Summaries
ames.describe()
```

```python
# Check for missing values
ames.isnull().sum()
```

```python
# value counts for categorical columns
categorical_cols = ames.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"Value counts for {col}:")
    print(ames[col].value_counts())
    print("\n")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.boxplot(ames["SalePrice"], vert=True, patch_artist=True)
plt.title("Distribution of Sale Prices")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.boxplot(ames["Lot Area"], vert=True, patch_artist=True)
plt.title("Distribution of Lot Area")
plt.ylabel("Lot Area")
plt.show()
```

### Exploratory Data Analysis Summary

The dataset includes 2,930 homes with a wide range of property and sale characteristics.  
The average sale price is $180,796, with a standard deviation of almost $80,000, showing high variability across houses.

**Numerical features:**
- Sale Price and Lot Area are both right-skewed, with a few high-value outliers.  
- Most homes were built between 1950 and 2000, and many have been remodeled since.  

**Categorical features:**
- The majority of homes fall under Residential Low Density (RL) zoning and have paved streets.  
- Neighborhoods like NAmes, CollgCr, and OldTown are the most common.  
- Most homes are single-family with poured concrete foundations, gas heating, and central air.  
- Garage and exterior quality ratings are typically average to good.  

**Sale details:**  
Most transactions are standard warranty deeds and normal sales, with only a few new or abnormal sales.  

**Key takeaways:**  
Distributions of Sale Price and Lot Area are heavily skewed, suggesting transformations (like log scaling) before regression.  
Many features show little variation (e.g., utilities, street type), while neighborhood and property size appear to drive much of the observed price differences.
